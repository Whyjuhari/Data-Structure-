```python
# Tipe data dictionary
# Yaitu tipe data array dimana key atau index dari array bisa berbentuk string,
# Tidak hanya number saja.

# Cara pembuatan tipe data dictionary
# Kita menggunakan tanda kurung kurawal.
# Selan=in itu setiap element merupakan pasangan dari key dan value.
# Antar satu element dengan element lain dipisah dengan tanda koma.
# Contoh sebagai berikut : 

nama_hero = {1: "Lancelot", 2: "Ling", 3: "Paquito"}
print(nama_hero)

# Kita juga bisa menampilkan hasil dari tipe data dictionary,
# Menggunakan perulangan for

print()
for hero in nama_hero:
    print(hero, ':', nama_hero[hero])


# Nilai atau value dari setiap element dictionary 
# bisa terdiri dari berbagai tipe data lain.
print()
my_data_dict = {"Kegiatan" : "Belajar",
                1: "Python", 2: "Css", 3: "Html",
                "menyerah"   : False,
                "target"     : 2024
                }
print(my_data_dict)

# Cara mengakses element dictionary python
print()
my_hero = {"Assasin" : "Lancelot",
           "Fighter" : "Paquito",
           "Tank"    : "Franko"
          }
print(my_hero["Fighter"])

# Cara mengubah element dictionary python
print()
my_hero = {"Assasin" : "Lancelot",
           "Fighter" : "Paquito",
           "Tank"    : "Franko"
          }
print(f"Sebelum diubah : {my_hero}")
my_hero["Tank"] = "Jhonson"
print(f"Sesudah diubah : {my_hero}")

# Cara menambah element dictionary python
print()
my_hero = {"Assasin" : "Lancelot",
           "Fighter" : "Paquito",
           "Tank"    : "Franko"
          }
print(f"Sebelum ditambahkan : {my_hero}")
my_hero["Mage"] = "Lylia"
print(f"Sesudah ditambahkan : {my_hero}")

# Cara menghapus element dictionary python
print()
my_hero = {"Assasin" : "Lancelot",
           "Fighter" : "Paquito",
           "Tank"    : "Franko"
          }
print(f"Sebelum Dihapus : {my_hero}")
del my_hero["Fighter"] 
print(f"Sesudah Dihapus : {my_hero}")
```
